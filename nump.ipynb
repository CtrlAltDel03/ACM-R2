```python
#Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates
import numpy as np
import math as m
rnd = np.random.default_rng()
a=rnd.random((10,2))
print("Cartesian coordinates : \n",a)
i=0
l=[]
while i<10 :
    r2=(a[i][0]*a[i][0])+(a[i][1]*a[i][1])
    r=m.sqrt(r2)
    s=(a[i][1]/a[i][0])
    th=m.atan(s)
    l2=[r,th]
    l.append(l2)
    i=i+1
print("Polar : ",l)
```
